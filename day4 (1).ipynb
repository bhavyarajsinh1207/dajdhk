1. Conditional Statements in Python
ðŸ”¹ Definition:

Conditional statements are used to make decisions in Python.
They execute certain blocks of code depending on whether a condition is True or False.

ðŸ”¹ Types:

if statement

if...else statement

if...elif...else statement

Nested if

Example 1: Simple if Statement

```python
age = 20
if age >= 18:
    print("You are eligible to vote.")
```

```python
You are eligible to vote.
```

Example 2: if...else Statement

```python
num = 5
if num % 2 == 0:
    print("Even number")
else:
    print("Odd number")
```

Odd number

Example 3: if...elif...else Statement

```python
marks = 75
if marks >= 90:
    print("Grade: A")
elif marks >= 70:
    print("Grade: B")
else:
    print("Grade: C")
```

Grade: B

Example 4: Nested if

```python
num = 10
if num > 0:
    if num % 2 == 0:
        print("Positive Even number")
```

Positive Even number

3. Pattern Printing in Python

Pattern printing uses loops to create shapes like triangles, squares, etc.

Example 1: Right Triangle Pattern

```python
for i in range(1, 6):
    print("*" * i)
```

Example 2: Inverted Triangle

```python
for i in range(5, 0, -1):
    print("*" * i)
```

Example 3: Pyramid Pattern

```python
rows = 5
for i in range(1, rows + 1):
    print(" " * (rows - i) + "*" * (2*i - 1))
```

Example 4: Number Pattern

```python
for i in range(1, 6):
    for j in range(1, i + 1):
        print(j, end=" ")
    print()
```
